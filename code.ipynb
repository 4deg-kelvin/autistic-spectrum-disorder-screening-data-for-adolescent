```python
# do label encoding
from sklearn.preprocessing import LabelEncoder
ethnicity_encoder = LabelEncoder()
country_encoder = LabelEncoder()
age_encoder = LabelEncoder()
relation_encoder = LabelEncoder()
asd_encoder = LabelEncoder()
juncide_encoder = LabelEncoder()
prev_family_encoder = LabelEncoder()


# fit and transformm all the columns above
merged_money_poverty_df['ethnicity'] = ethnicity_encoder.fit_transform(merged_money_poverty_df['ethnicity'])
merged_money_poverty_df['contry_of_res'] = country_encoder.fit_transform(merged_money_poverty_df['contry_of_res'])
merged_money_poverty_df['age_desc'] = age_encoder.fit_transform(merged_money_poverty_df['age_desc'])
merged_money_poverty_df['relation'] = relation_encoder.fit_transform(merged_money_poverty_df['relation'])
merged_money_poverty_df['Class/ASD'] = asd_encoder.fit_transform(merged_money_poverty_df['Class/ASD'])
merged_money_poverty_df['jundice'] = juncide_encoder.fit_transform(merged_money_poverty_df['jundice'])


```
